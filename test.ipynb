```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
cols = ['SPEED_A','TORQUE_A','P_OILOF']
read = pd.read_excel('VL.xlsx',usecols=cols,skiprows=[1],header=0)
x=read.SPEED_A
y=read.TORQUE_A
z=read.P_OILOF
fig=plt.figure()
ax=Axes3D(fig)
ax.plot3D(x,y,z)
plt.show()
```
